```python
import numpy as np
import matplotlib.pyplot as plt
def plot_quadratic_equation(a, b, c):
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        root1 = (-b + np.sqrt(discriminant)) / (2*a)
        root2 = (-b - np.sqrt(discriminant)) / (2*a)
        plt.scatter([root1, root2], [0, 0], color='red', label='Roots')
        plt.annotate(f'({root1:.2f}, 0)', xy=(root1, 0), xytext=(root1, 10), arrowprops=dict(arrowstyle='->'))
        plt.annotate(f'({root2:.2f}, 0)', xy=(root2, 0), xytext=(root2, 10), arrowprops=dict(arrowstyle='->'))
    if a > 0:
        vertex_x = -b / (2*a)
        vertex_y = a * vertex_x**2 + b * vertex_x + c
        plt.scatter(vertex_x, vertex_y, color='green', label='Vertex')
        plt.annotate(f'({vertex_x:.2f}, {vertex_y:.2f})', xy=(vertex_x, vertex_y), xytext=(vertex_x-2, vertex_y+10), arrowprops=dict(arrowstyle='->'))
        plt.legend()
    plt.title('Quadratic Equation')
    plt.grid(True)
    plt.show()
plot_quadratic_equation(-3, 4, 7)
plot_quadratic_equation(5, -4, 9)
```
